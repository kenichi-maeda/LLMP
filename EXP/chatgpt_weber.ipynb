```python
image, label = L.GPImage.weber('10')
imshow(image, cmap='gray')

question = 'What do you see? If you see dots, then proceed as follows: Estimate the number of black dots in the image. It does not have to be precise; just tell me your estimation.'
answer = L.ChatGPT.query(question, image)
print(answer, 'GT:', label)
```

```python
import re
matches = re.search(r'\d+', answer)
if matches:
    number = int(matches.group())
    error = number - 10 - label
    print(error)
```
